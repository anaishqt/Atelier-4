```java
import java.util.Scanner;

// --- 1. LES FONCTIONS D'AFFICHAGE DU TABLEAU (Étape 1) ---

void displayGridLine() {
    System.out.println("+---------------+---------+-----+----------+");
}

void displayProductGridHeader() {
    displayGridLine();
    System.out.println("| Product       | € Unit  | Qty |   price  |");
    displayGridLine();
}

void displayProductGridFooter(double total) {
    displayGridLine();
    System.out.printf("                                | %7.2f€ |\n", total);
    System.out.println("                                +----------+");
}

// --- 2. LA NOUVELLE FONCTION DE SAISIE (Étape 2) ---

int inputProductQuantity(Scanner keyboard, String productName, double unitPrice) {
    int quantity;
    // La boucle do...while pour répéter si erreur
    do {
        System.out.printf("How many %s at €%.2f ?\n", productName, unitPrice);
        quantity = keyboard.nextInt();
        
        // Affichage du message d'erreur si ce n'est pas entre 0 et 900
        if (quantity < 0 || quantity > 900) {
            System.out.println("Erreur : La quantité doit être comprise entre 0 et 900. Veuillez recommencer.");
        }
        
    } while (quantity < 0 || quantity > 900); // On boucle tant que c'est faux
    
    return quantity; // On renvoie la quantité correcte
}

// --- PROGRAMME PRINCIPAL COMMENCE ICI ---

Scanner keyboard = new Scanner(System.in);

String productName1 = "Stylos"; 
double unitPrice1 = 1.2; 
// Appel de la nouvelle fonction pour les stylos !
int quantity1 = inputProductQuantity(keyboard, productName1, unitPrice1);

String productName2 = "Cahier";
double unitPrice2 = 3.5;
// Appel de la nouvelle fonction pour les cahiers !
int quantity2 = inputProductQuantity(keyboard, productName2, unitPrice2);

double total = 0;

displayProductGridHeader();

double price1 = unitPrice1 * (double)quantity1;
total += price1;
System.out.printf("| %-13s | %6.2f€ | %3d | %7.2f€ |\n", productName1, unitPrice1, quantity1, price1);

double price2 = unitPrice2 * (double)quantity2;
total += price2;
System.out.printf("| %-13s | %6.2f€ | %3d | %7.2f€ |\n", productName2, unitPrice2, quantity2, price2);

displayProductGridFooter(total);
```

```java
import java.util.Scanner;

// --- 1. FONCTIONS D'AFFICHAGE (Étape 1) ---
void displayGridLine() {
    System.out.println("+---------------+---------+-----+----------+");
}

void displayProductGridHeader() {
    displayGridLine();
    System.out.println("| Product       | € Unit  | Qty |   price  |");
    displayGridLine();
}

void displayProductGridFooter(double total) {
    displayGridLine();
    System.out.printf("                                | %7.2f€ |\n", total);
    System.out.println("                                +----------+");
}

// --- 2. FONCTION DE SAISIE (Étape 2) ---
int inputProductQuantity(Scanner keyboard, String productName, double unitPrice) {
    int quantity;
    do {
        System.out.printf("How many %s at €%.2f ?\n", productName, unitPrice);
        quantity = keyboard.nextInt();
        if (quantity < 0 || quantity > 900) {
            System.out.println("Erreur : La quantité doit être comprise entre 0 et 900. Veuillez recommencer.");
        }
    } while (quantity < 0 || quantity > 900);
    return quantity;
}

// --- 3. NOUVELLE FONCTION POUR LA LIGNE PRODUIT (Étape 3) ---
double displayProductGridEntry(String productName, double unitPrice, int quantity) {
    // 1. On calcule le prix total de cette ligne
    double rowPrice = unitPrice * (double)quantity;
    
    // 2. On affiche la belle ligne du tableau
    System.out.printf("| %-13s | %6.2f€ | %3d | %7.2f€ |\n", productName, unitPrice, quantity, rowPrice);
    
    // 3. On RETOURNE le prix calculé 
    return rowPrice;
}

// --- PROGRAMME PRINCIPAL COMMENCE ICI ---

Scanner keyboard = new Scanner(System.in);

String productName1 = "Stylos"; 
double unitPrice1 = 1.2; 
int quantity1 = inputProductQuantity(keyboard, productName1, unitPrice1);

String productName2 = "Cahier";
double unitPrice2 = 3.5;
int quantity2 = inputProductQuantity(keyboard, productName2, unitPrice2);

// Affichage de l'en-tête
displayProductGridHeader();

// On prépare le total à 0
double total = 0;


total += displayProductGridEntry(productName1, unitPrice1, quantity1);
total += displayProductGridEntry(productName2, unitPrice2, quantity2);

// Affichage de la fin du tableau avec le total
displayProductGridFooter(total);
```

## BONUS


```java
import java.util.Scanner;


void displayGridLine() {
    System.out.println("+---------------+---------+-----+----------+");
}

void displayProductGridHeader() {
    displayGridLine();
    System.out.println("| Product       | € Unit  | Qty |   price  |");
    displayGridLine();
}

void displayProductGridFooter(double total) {
    displayGridLine();
    System.out.printf("                                | %7.2f€ |\n", total);
    System.out.println("                                +----------+");
}

int inputProductQuantity(Scanner keyboard, String productName, double unitPrice) {
    int quantity;
    do {
        System.out.printf("How many %s at €%.2f ?\n", productName, unitPrice);
        quantity = keyboard.nextInt();
        if (quantity < 0 || quantity > 900) {
            System.out.println("Erreur : La quantité doit être comprise entre 0 et 900. Veuillez recommencer.");
        }
    } while (quantity < 0 || quantity > 900);
    return quantity;
}

double displayProductGridEntry(String productName, double unitPrice, int quantity) {
    double rowPrice = unitPrice * (double)quantity;
    System.out.printf("| %-13s | %6.2f€ | %3d | %7.2f€ |\n", productName, unitPrice, quantity, rowPrice);
    return rowPrice;
}

// --- NOUVEAU PROGRAMME PRINCIPAL (BONUS) ---

Scanner keyboard = new Scanner(System.in);

// 1. On crée des "Tableaux" (listes) avec plein de produits 
String[] productNames = {"Stylos", "Cahier", "Gomme", "Classeur", "Sac à dos"};
double[] unitPrices = {1.20, 3.50, 0.90, 4.50, 25.00};
int[] quantities = new int[productNames.length]; // Une liste vide pour stocker les réponses

// 2. On utilise une BOUCLE 'for' pour poser la question pour chaque produit de la liste
System.out.println("--- BIENVENUE DANS LA BOUTIQUE ---");
for (int i = 0; i < productNames.length; i++) {
    quantities[i] = inputProductQuantity(keyboard, productNames[i], unitPrices[i]);
}

// 3. On prépare l'affichage
System.out.println("\n--- VOTRE TICKET DE CAISSE ---");
displayProductGridHeader();
double total = 0;

// 4. On utilise une autre BOUCLE pour dessiner les lignes du tableau
for (int i = 0; i < productNames.length; i++) {
    // on n'affiche la ligne QUE si la quantité est supérieure à 0 !
    if (quantities[i] > 0) {
        total += displayProductGridEntry(productNames[i], unitPrices[i], quantities[i]);
    }
}

// 5. On affiche la fin avec le grand total
displayProductGridFooter(total);
```
